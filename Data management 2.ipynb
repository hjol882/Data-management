```python
import csv
from collections import defaultdict
from datetime import datetime
```

```python
data = defaultdict(lambda: defaultdict(list)) # initialize a nested dictionary whith  productAggrupationID
january = defaultdict(list) # initialize a nested dictionary ordered in january  whith productAggrupationID


def clean_product_id(product_id): # convert the productID to a string
    try:
        return str(int(product_id))
    except ValueError:
        try:
            return str(int(float(product_id)))
        except ValueError:
            return product_id

def unique(list1): # return a sorted list of the unique elements
     return sorted(set(list1))  

with open("C:\\Users\sampleData.csv", mode='r') as file: # Open the cvs in read mode
    reader = csv.DictReader(file)

    all_product_agrupation_ids = [] # will store all the  productAggrupationID
    all_customer_ids = [] # will store all the  Customers
    repetition = {}

    for row in reader:  # extracts all the values of the rows of every column 
        customer = row['customer']
        productAggrupationID = clean_product_id(row['productAgrupationID'])
        productID = int(row['productID'])
        order_date = datetime.strptime(row['order_date'], "%Y-%m-%dT%H:%M:%S.%fZ")
        product_cases = float(row['product_cases_ordered'])
   
        all_product_agrupation_ids.append(productAggrupationID) # append the productAggrupationID
        all_customer_ids.append(customer) # # append the productAggrupationID into a list

        if order_date.month == 1: # check if order.date is in january
          january[customer].append(productAggrupationID)

    data[customer][productAggrupationID].append({ # appends a dictionary to the list asociate to customer and productAggrupationID
        'order_date': order_date,
        })
    
    
    unique_product_agrupation_ids = unique(all_product_agrupation_ids) # set the unique values of all_product_agrupation_ids
     
    unique_customer_ids = unique(all_customer_ids) # set the unique values of all_product_agrupation_ids


    for i in all_product_agrupation_ids: # count the repetions of the all_product_agrupation_ids
        if i in   repetition :
          repetition[i] += 1
        else:
           repetition[i] = 0

    print( repetition)



print("Product IDs ordered in January per customer:")
for customer, product_ids in january.items(): # loop throught each custormer and their product_ids
    sorted_unique_product_ids = sorted(product_ids) # sorted the prodcut_ids
    print(f"Customer: {customer}") # print the customers
    print(f"  Product IDs: {sorted_unique_product_ids}") # print the product_ids associate to the customer
```
